### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
tienda.info()
```

```python
ingresos_totales = tienda['Precio'].sum()
ingresos_totales2 = tienda2['Precio'].sum()
ingresos_totales3 = tienda3['Precio'].sum()
ingresos_totales4 = tienda4['Precio'].sum()

print(f'Ingresos totales para la primera tienda: {ingresos_totales}')
print(f'Ingresos totales para la segunda tienda: {ingresos_totales2}')
print(f'Ingresos totales para la tercera tienda: {ingresos_totales3}')
print(f'Ingresos totales para la cuarta tienda:  {ingresos_totales4}')
```

```python
print('La primera tienda tiene mayor cantidad de ingresos')
```

# 2. Ventas por categoría

```python
tienda_categoria = tienda['Categoría del Producto'].value_counts()
tienda2_categoria = tienda2['Categoría del Producto'].value_counts()
tienda3_categoria = tienda3['Categoría del Producto'].value_counts()
tienda4_categoria = tienda4['Categoría del Producto'].value_counts()

print(f'Ventas por categoria de la primera tienda:\n{tienda_categoria}')
print(f'Ventas por categoria segunda tienda:\n{tienda2_categoria}')
print(f'Ventas por categoria tercera tienda:\n{tienda3_categoria}')
print(f'Ventas por categoria cuarta tienda:\n{tienda4_categoria}')
```

# 3. Calificación promedio de la tienda


```python
calificacion_tienda = float(tienda['Calificación'].mean())
calificacion_tienda2 = float(tienda2['Calificación'].mean())
calificacion_tienda3 = float(tienda3['Calificación'].mean())
calificacion_tienda4 = float(tienda4['Calificación'].mean())
```

```python
print(f'La calificacion promedio de los usuarios para la primera tienda es :{calificacion_tienda:.2f}')
print(f'La calificacion promedio de los usuarios para la segunda tienda es :{calificacion_tienda2:.2f}')
print(f'La calificacion promedio de los usuarios para la tercera tienda es :{calificacion_tienda3:.2f}')
print(f'La calificacion promedio de los usuarios para la cuarta tienda es :{calificacion_tienda4:.2f}')
```

# 4. Productos más y menos vendidos

```python
max_tienda = tienda['Producto'].value_counts().idxmax()
max_tienda2 = tienda2['Producto'].value_counts().idxmax()
max_tienda3 = tienda3['Producto'].value_counts().idxmax()
max_tienda4 = tienda4['Producto'].value_counts().idxmax()
```

```python
min_tienda = tienda['Producto'].value_counts().idxmin()
min_tienda2 = tienda2['Producto'].value_counts().idxmin()
min_tienda3 = tienda3['Producto'].value_counts().idxmin()
min_tienda4 = tienda4['Producto'].value_counts().idxmin()
```

```python
print(f'El producto mas vendido de la primera tienda fue el {max_tienda} y el menos vendido fue {min_tienda}.')
print(f'El producto mas vendido de la segunda tienda fue {max_tienda2} y el menos vendido fue {min_tienda2}.')
print(f'El producto mas vendido en la tercera tienda fue {max_tienda3} y el menos vendido fue {min_tienda3}.')
print(f'El producto mas vendido en la cuarta tienda fue {max_tienda4} y el menos vendido fue {min_tienda4}.')
```

# 5. Envío promedio por tienda

```python
costenvio_tienda = tienda['Costo de envío'].mean()
constenvio_tienda2 = tienda2['Costo de envío'].mean()
costenvio_tienda3 = tienda3['Costo de envío'].mean()
costenvio_tienda4 = tienda4['Costo de envío'].mean()
```

```python
print(f'costo de envio promedio de la primera tienda :{costenvio_tienda:.3f}')
print(f'costo de envio promedio de la segunda tienda :{constenvio_tienda2:.3f}')
print(f'costo de envio promedio de la tercera tienda :{costenvio_tienda3:.3f}')
print(f'costo de envio promedio de la cuarta tienda :{costenvio_tienda4:.3f}')
```

```python
print('El cliente de la primera tienda paga mas por el envio del producto, y el que menos paga por envio es en el de la cuarta tienda')
```
